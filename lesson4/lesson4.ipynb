```python
t = np.linspace(0,10,51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show
```

Задание 3

```python
x = np.linspace(-3, 3, 51)
print(x)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim(-5, 5)
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
```

```python
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv('creditcard.csv')
```

```python
print(dataset)
```

```python
info = dataset['Class'].value_counts()
info.plot(kind='bar')
```

```python
info.plot(kind='bar', logy=True)
```

```python
v10 = dataset['V1'].loc[dataset['Class'] == 0]
v11 = dataset['V1'].loc[dataset['Class'] == 1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.subplots_adjust(wspace=1)

ax[0].hist(v10, density=True, bins=20, color='grey', alpha=0.5)
ax[0].legend(['Class 0'])
ax[1].hist(v11, density=True, bins=20, color='red', alpha=0.5)
ax[1].legend(['Class 1'])
```
